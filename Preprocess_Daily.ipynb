```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./original.csv')
```

```python
df['Volume_(BTC)'] = df['Volume_(BTC)'].fillna(0)
df['Volume_(Currency)'] = df['Volume_(Currency)'].fillna(0)


df.Timestamp = pd.to_datetime(df.Timestamp, unit='s')
df['Year'] = df['Timestamp'].dt.year


```

```python
df = df[df.Year != 2011]
df = df[df.Year != 2012]
df = df[df.Year != 2013]
df.head()
```

```python
BTC_Volume = []
BTC_count = 0
BTC_element = 0
for BTC_volume in ((df['Volume_(BTC)'])):
    BTC_count += 1
    BTC_element += BTC_volume
    BTC_Volume.append(BTC_element)
    if BTC_count == 59:
        BTC_count = 0
        BTC_element = 0

CURRENCY_Volume = []
CURRENCY_count = 0
CURRENCY_element = 0
for CURRENCY_volume in ((df['Volume_(Currency)'])):
    CURRENCY_count += 1
    CURRENCY_element += CURRENCY_volume
    CURRENCY_Volume.append(CURRENCY_element)
    if CURRENCY_count == 59:
        CURRENCY_count = 0
        CURRENCY_element = 0
        
```

```python
df= df.reset_index()
df.head()
```

```python
to_be_added1 = pd.DataFrame({'BTC_Volume':BTC_Volume})
df = pd.concat([df, to_be_added1], axis=1)
to_be_added2 = pd.DataFrame({'CURRENCY_Volume':CURRENCY_Volume})
df = pd.concat([df, to_be_added2], axis=1)
df.head(20)
```

```python
df['Open'] = df['Open'].fillna(0)
df = df[df['Open'] != 0.000000]
```

```python
df= df.reset_index()
df.head()
```

```python
Opens = []
open_count = 0
open_element = 0
df['level_0']= df['level_0'].astype(int)

for opens in ((df['Open'])):
    if int(df.iat[open_count,0]%1440)==0:
        open_element=opens 
    Opens.append(open_element)
    open_count += 1
to_be_added3 = pd.DataFrame({'Daily_Open':Opens})
df = pd.concat([df, to_be_added3], axis=1)
df.head(1400)
```

```python
Closes = []
close_count = 0
close_element = 0

for closes in ((df['Close'])):
    if int(df.iat[close_count,0]%1440)==1439:
        close_element=closes 
    Closes.append(close_element)
    close_count += 1
to_be_added4 = pd.DataFrame({'Daily_Close':Closes})
df = pd.concat([df, to_be_added4], axis=1)
df.head(1300)
```

```python
Highs = []
high_count = 0
high_element = 0
for highs in ((df['High'])):
    if int(df.iat[high_count,0]%1440)==0:
        high_element=highs
    if highs>high_element:
        high_element=highs
    Highs.append(high_element)
    high_count += 1
to_be_added5 = pd.DataFrame({'Daily_High':Highs})
df = pd.concat([df, to_be_added5], axis=1)
df.head(1000)
```

```python
Lows = []
low_count = 0
low_element = 0
for lows in ((df['Low'])):
    if int(df.iat[low_count,0]%1440)==0:
        low_element=lows
    if lows<low_element:
        low_element=lows
    Lows.append(low_element)
    low_count += 1
to_be_added6 = pd.DataFrame({'Daily_Low':Lows})
df = pd.concat([df, to_be_added6], axis=1)
df.head()
```

```python
df.drop('level_0', inplace=True, axis=1)
df.drop('index', inplace=True, axis=1)
df.drop('Open', inplace=True, axis=1)
df.drop('Close', inplace=True, axis=1)
df.drop('High', inplace=True, axis=1)
df.drop('Low', inplace=True, axis=1)
df.drop('Volume_(BTC)', inplace=True, axis=1)
df.drop('Volume_(Currency)', inplace=True, axis=1)
df.drop('Weighted_Price', inplace=True, axis=1)
df.drop('Year', inplace=True, axis=1)
```
